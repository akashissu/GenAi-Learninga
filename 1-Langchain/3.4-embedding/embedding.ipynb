```python
###Embedding Technique
Converting Text into a vector is called embedding. It is a numerical representation of text that captures its semantic meaning. This allows us to perform various operations on the text, such as searching, clustering, and classification. In this notebook, we will learn how to create embeddings using OpenAI API and store them in a vector database for efficient retrieval.
```

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ['OPENAI_API_KEY']=os.getenv('OPENAI_API_KEY')
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text="This is a sample text to create an embedding vector With the help of OpenAI API"
query_result = embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
embeddings_1024 = OpenAIEmbeddings(model="text-embedding-3-large",dimensions=1024)
embeddings_1024
```

```python
text="This is a sample text to create an embedding vector With the help of OpenAI API"
query_result = embeddings_1024.embed_query(text)
len(query_result)
```

```python
len(query_result)
```

```python
##Vector Embedding And Vector StoreDatabase
#Vector databases are designed to store and manage high-dimensional vectors, such as those generated by embedding techniques. They provide efficient storage and retrieval of vectors, allowing for fast similarity searches and other operations. In this section, we will learn how to use a vector database to store and retrieve embeddings.
```

```python
##Load and Split Documents
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

loader = PyPDFLoader("/Users/akash/Desktop/Langchain/SudhaResearch.pdf")
docs = loader.load()

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
final_documents = text_splitter.split_documents(docs)
final_documents
```

```python
from langchain_community.vectorstores import Chroma

db=Chroma.from_documents(final_documents, embeddings_1024)
db
```
